## Low Resolution Image Generation

```python
from PIL import Image
from glob import glob
import os
```

```python
path_splitter = '\\'
#base_path = 'C:\\Users\\ziipp\\Downloads\\DIV2K\\DIV2K_train_HR\\'
base_path = 'C:\\Users\\ziipp\\Downloads\\DIV2K_Flicker2K_DF2K\\DIV2K_valid_HR\\'

target_scale = 4
target_path = 'C:\\Users\\ziipp\\Downloads\\DIV2K_Flicker2K_DF2K\\DIV2K_valid_LR_bicubic_X' + str(target_scale) + '\\'

os.makedirs(target_path, exist_ok=True)

print(base_path)
print(target_path)
```

```python
file_list = glob(base_path + '*.png')
# file_list
```

```python
for file in file_list:
    filename = file.split(path_splitter)[-1]
    # print(filename)
    #
    img = Image.open(file)
    
    target_width = int(img.width/target_scale)
    target_height = int(img.height/target_scale)

    img_resize = img.resize((target_width, target_height), Image.BICUBIC)
    img_resize.save(target_path+filename)

    # img_resize_lanczos = img.resize((256, 256), Image.LANCZOS)
    # img_resize_lanczos.save('./test/hr_resize/'+filename)
    print(f'{filename} ({target_width}, {target_height})')
    
print('done')
```

```python
train_file_list = glob('C:\\Users\\ziipp\\Downloads\\DIV2K_Flicker2K_DF2K\\DF2K_train_HR\\*.png')

meta_filename = 'C:\\Users\\ziipp\\Downloads\\DIV2K_Flicker2K_DF2K\\DF2K_train_HR_info.txt'
f = open(meta_filename, 'w')

for file in train_file_list:
    filename = file.split('\\')[-1]
    # print(filename)
    
    img = Image.open(file)
    width = str(img.width)
    height = str(img.height)
    
    meta = filename + ' (' + width + ',' + height + ') 1\n'
    f.write(meta)
    
f.close()
    
print('done')
```

```python
from glob import glob
import os
import shutil
import random

base_paths = ['E:\\dl_data\\SR\\open_sr\\train\\hr\\', 
             'E:\\dl_data\\SR\\open_sr\\train\\lr\\']
new_train_paths = ['E:\\dl_data\\SR\\open_sr\\train\\hr_split_train\\',
                  'E:\\dl_data\\SR\\open_sr\\train\\lr_split_train\\']
new_valid_paths = ['E:\\dl_data\\SR\\open_sr\\train\\hr_split_valid\\',
                  'E:\\dl_data\\SR\\open_sr\\train\\lr_split_valid\\']

random.seed(2023)

for index, base_path in enumerate(base_paths):
    new_train_path = new_train_paths[index]
    new_valid_path = new_valid_paths[index]
    
    print(new_train_path)
    print(new_valid_path)
    
    shutil.rmtree(new_train_path, ignore_errors=True)
    os.makedirs(new_train_path, exist_ok=True)
    shutil.rmtree(new_valid_path, ignore_errors=True)
    os.makedirs(new_valid_path, exist_ok=True)

    file_list = glob(base_path + '*.png')
    file_count = len(file_list)
    print(file_count)
    
    file_list.sort()

    #random.shuffle(file_list)

    train_ratio = 0.975
    train_index = int(file_count * train_ratio)
    print(train_index)
    print(file_count - train_index)

    for file in file_list[:train_index]:
        filename = file.split('\\')[-1]
        shutil.copy(file, new_train_path+filename)
        #print(filename)    
    print('split for train done')

    for file in file_list[train_index:]:
        filename = file.split('\\')[-1]
        shutil.copy(file, new_valid_path+filename)
        #print(filename)
    print('split for valid done')
    
print('all done')
```
